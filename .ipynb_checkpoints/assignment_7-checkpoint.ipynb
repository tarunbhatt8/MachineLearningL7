# Exploratory data analysis:
## Titanic Disaster
The data set consists of 889 passengers who rode aboard the Titanic (total 2224).
(training set available from Kaggle.com)

The sinking of RMS Titanic is one of the most infamous shipwrecks in history.
On April 15, 1912, during her maiden voyage, the Titanic sank after coliding with an iceberg, killing 1502 out of 2224 passengers and crew.
The senational tragedy shocked the international community and led to better safety regulations for ships.
A movie  on this theme, directed by famous director James Cameron was released in 1997, an perhaps most of us only know about the tragedy from there only. Here a modern day Data Analysis can largely help you understand the tragedy in a far better manner in one go, than daily Internet Memes, so go through the EDA below patiently, perhaps you'll find something Interesting.
<img src="https://image.ibb.co/mWZyqJ/2.png" alt="2" border="0">

One of the reasons that the shipwreck lead to such loss of life was that there were not enough lifeboats for the passengers and crew. although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and upper-class.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset.
df = pd.read_csv("titanic.csv")
```

```python
#Printing first 5 columns from our data frame
df.head()
```

```python
#Observing the lables of each column
print(df.keys())
```

```python
#Observing the shape of our data frame.
df.shape
```

```python
#Lets check for missing values
#there seem to be no missing values present in our dataset
df.info()

##Displaying the total of NaN values in each column
df.isnull().sum()
```

```python
df.describe()
```

## Univariate Analysis

```python
df.Survived.value_counts()
#shows the number of people who survived and didn't survive
```

```python
# Distplot to show the survivors
sns.distplot(df['Survived'])
```

```python
# Histogram to show the gender distribution abord
df.Sex.value_counts().plot(kind='bar')
```

```python
# Distplot to show the class distribution abord
sns.distplot(df['Pclass'])
```

```python
# Statistical Values for all people aboard as per age
df.Age.describe()
```

```python
#box-plot to show agewise distribution of passengers onboard
sns.boxplot(data=df['Age'])
```

```python
# AGE-WISE DISTRIBUTION OF PASSENGERS ONBOARD
# see there are certain lucky old people who survived, can be considered to be outliers
# outliers are generally handled by considering the values only between 25 to 75 percentiles 
# Here the distribution seems to be skewed in terms of normal distribution
df.Age.hist(bins=20)
```

## Bivariate Analysis

```python
df.corr(method='pearson')
```

```python
# an analysis of survivors catgorised by class
sns.factorplot('Survived', col='Pclass', data=df, kind='count', size=7, aspect=.8)
```

```python
# so which class sufferd more casulities comaritively???
# factors responsible: Go watch the movie YOU'LL KNOW 'WHY' 
sns.countplot(x='Survived', data=df, hue='Pclass')
```

```python
#an analysis of survivors catgorised by Siblings/Spouses Aboard
sns.factorplot('Siblings/Spouses Aboard', col='Survived', data=df, kind='count', size=7, aspect=.8)
```

```python
#an analysis of survivors catgorised by Parents/Children Aboard
sns.factorplot('Parents/Children Aboard', col='Survived', data=df, kind='count', size=7, aspect=.8)
```

```python
# which age group spent more on the fare?
sns.jointplot(x='Age', y='Fare',\
              data=df, kind='scatter')
```

## Multivariate Analysis

#### Interesting Graph from the Analysis

```python
#Noticed some Interesting Insights?? 
#Survivors mostly were accompanied by thier parent/children.
#You already know the reason, most of the boats were first filled by Women & Children only, as per Captain's early Orders.
sns.factorplot(x='Pclass', y='Parents/Children Aboard', col='Survived', \
              data=df, kind='box', col_wrap=4, size=3)
```

#### Third Interesting Graph from the analysis

```python
#Interesting Insight??
# Somehow people with no Siblings/Spouses Aboard and belonging to First Class died a little less than others.
sns.factorplot(x='Pclass', y='Siblings/Spouses Aboard', col='Survived', \
              data=df, kind='box', col_wrap=4, size=3)
```
